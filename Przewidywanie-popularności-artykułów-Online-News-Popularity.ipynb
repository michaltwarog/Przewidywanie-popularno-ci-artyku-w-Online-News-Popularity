```python
sort_shares = df.sort_values('shares') 
median_shares = sort_shares['shares'].median() 
q1_shares  = sort_shares['shares'].quantile(q=0.25) 
q3_shares  = sort_shares['shares'].quantile(q=0.75) 
iqr_shares = q3_shares - q1_shares
```

```python
in_bound1 = q1_shares-(iqr_shares*1.5) 
print("dolna obserwacja odstająca = ",in_bound1)
in_bound2 = q3_shares+(iqr_shares*1.5)
print("górna obserwacja odstająca = ",in_bound2)
out_bound1 = q1_shares-(iqr_shares*3)
print("dolna obserwacja ekstreamalna = ", out_bound1)
out_bound2 = q3_shares+(iqr_shares*3)
print("górna obserwacja ekstremalna = ", out_bound2)
```

```python
data_np = np.array(data, dtype=np.int16)
target_np = np.array(target, dtype=np.int16)

print(data_np.shape)
print(target_np.shape)
```

```python
print("Pierwszy artykuł z bazy danych")
print(data_np[1,:])
print('---Mean---')
print(data_np.mean(axis=0))
print('--std---')
print(data_np.std(axis=0))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_np)
```

```python
print("Pierwszy artykuł z bazy danych")
print(scaled_data[1,:])
print('---Mean---')
print(scaled_data.mean(axis=0))
print('--std---')
print(scaled_data.std(axis=0))
```

```python
from sklearn.model_selection import train_test_split
news_train_data, news_test_data, \
news_train_target, news_test_target = \
train_test_split(scaled_data, target_np, test_size=0.1, random_state=40)
```

```python
print("Zbiór trenujący:")
print("news_train_data:", news_train_data.shape)
print("news_train_target:", news_train_target.shape)
```

```python
print("Zbiór uczący:")
print("news_test_data:", news_test_data.shape)
print("news_test_target:", news_test_target.shape)
```

LinearRegression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
linear_regression=LinearRegression()
scores = cross_val_score(LinearRegression(), data, target,cv=10)
print(scores)
print("Średnia wyników dla współczynnika determinacji: ",scores.mean())
```

DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
decision_tree_regressor= DecisionTreeRegressor()
scores = cross_val_score(decision_tree_regressor,news_train_data,news_train_target,scoring='r2',cv=10)
print(scores)
print("Średnia wyników dla współczynnika determinacji: ",scores.mean())
```

KNeighborsRegressor

```python
from sklearn.neighbors import KNeighborsRegressor
k_neighbors_regressor = KNeighborsRegressor()
scores=cross_val_score(k_neighbors_regressor,news_train_data,news_train_target,cv=10)
print(scores)
print("Średnia wyników dla współczynnika determinacji: ",scores.mean())
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import Normalizer


n_news_train_data, n_news_test_data, \
n_news_train_target, n_news_test_target = \
train_test_split(data_np, target_np, test_size=0.1, random_state=40)

pt = PolynomialFeatures(2, )
news_train_poly=pt.fit_transform(n_news_train_data)
news_test_poly=pt.fit_transform(n_news_test_data)

normalizer = Normalizer()
normalized_train_poly = normalizer.fit_transform(news_train_poly)
normalized_test_poly = normalizer.fit_transform(news_test_poly)

news_train_poly.shape
```

```python
def show_scores(type_of_regressor,news_train, news_test, name):
    type_of_regressor.fit(news_train, n_news_train_target)    
    print("Błąd średniokwadratowy wyuczonego modelu", name, ": %.2f" % 
      mean_squared_error(news_test_target, type_of_regressor.predict(news_test)))
    print('Współczynnik determinacji: %.2f' % type_of_regressor.score(news_test, n_news_test_target))
```

```python


print("Regresja liniowa:")
show_scores(linear_regression, news_train_poly, news_test_poly, "regresji liniowej dla przetransformowanych danych do cech wielomianowych")
show_scores(linear_regression, normalized_train_poly,normalized_test_poly, "regresji liniowej dla znormalizowanych oraz przetransformowanych do cech wielomianowych danych")

print("\nDTR:")
show_scores(decision_tree_regressor, news_train_poly, news_test_poly, "DTR dla przetransformowanych danych do cech wielomianowych")
show_scores(decision_tree_regressor, normalized_train_poly,normalized_test_poly, "DTR dla znormalizowanych oraz przetransformowanych do cech wielomianowych danych")

print("\nKNN:")
show_scores(k_neighbors_regressor, news_train_poly, news_test_poly, "KNN dla przetransformowanych danych do cech wielomianowych")
show_scores(k_neighbors_regressor, normalized_train_poly,normalized_test_poly, "KNN dla znormalizowanych oraz przetransformowanych do cech wielomianowych danych")
```
